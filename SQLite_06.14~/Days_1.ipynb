```python
#함수 

import  sqlite3

def trans_money(db_cursor, account_id, money_to_trans):
    Is_Successful = False
    
    try:
        strSQL = "INSERT INTO bank_transaction (trans_date, account_id, trans_amount ) " 
        strSQL = strSQL +  "values ('2021-06-11 13:00:00' ," 
        strSQL = strSQL + str(account_id) + ", "
        strSQL = strSQL + str(money_to_trans) + " ) "
        
        print(strSQL)
        
        db_cursor.execute(strSQL)
        Is_Successful = True
    except:
        Is_Successful = False
    
    return Is_Successful


def get_current_balance(db_cursor, account_id):
    strSQL = "SELECT account_id, sum(trans_amount) cur_balance FROM bank_transaction "
    strSQL = strSQL + " WHERE account_id = " + str(account_id)
    strSQL = strSQL + " GROUP BY account_id "
    cur.execute(strSQL)
    
    results = cur.fetchall()
    
    
    cur_balance = results[0][1]
    
    return cur_balance 


#--------------------------------------------------------------------------------------------------------

conn = sqlite3.connect('chinook.db')
cur = conn.cursor()
    

account_id_A = 20210001
account_id_B = 20210002
money_to_trans = 5000;

rtnTransA = trans_money(cur, account_id_A, -money_to_trans)
rtnTransB = trans_money(cur, account_id_B, money_to_trans)

if rtnTransA and rtnTransB:
    conn.commit()
    
    curBalanceA = get_current_balance(cur, account_id_A)    
    print(curBalanceA)
    
    curBalanceB = get_current_balance(cur, account_id_B)    
    print(curBalanceB)
    
else:
    conn.rollback()
    
    
curBalanceA = get_current_balance(cur, account_id_A)
# print(curBalanceA)



conn.close()    
    
```
